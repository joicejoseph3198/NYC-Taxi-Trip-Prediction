```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns
import plotly as py
import plotly.express as px
from plotly.offline import iplot, plot , init_notebook_mode
init_notebook_mode(connected=True)
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

```python
fr_1 = pd.read_csv('fastest_routes_train_part_1.csv')
fr_2 = pd.read_csv('fastest_routes_train_part_2.csv')
fr = pd.concat([fr_1,fr_2])

fr_new = fr[['id', 'total_distance', 'total_travel_time', 'number_of_steps']]


train_df = pd.read_csv('train.csv')
train = pd.merge(train_df,fr_new, on = 'id', how = 'left')

train_df = train.copy()

train_df['DistanceKM']= (train_df['total_distance']/1000)
train_df['DurationMIN']= (train_df['trip_duration']/60)

test_df = pd.read_csv('test.csv') 
```

```python
train_df.isnull().sum()
#there were 3 missing values
train_df.dropna(axis = 0, how = 'any',inplace =True)
```

```python
train_df.info()
```

```python
#removing outliers
distLimits = [np.percentile(train_df['total_distance'], 0.3), np.percentile(train_df['total_distance'],99.7)]
durLimits  = [np.percentile(train_df['trip_duration'], 0.4), np.percentile(train_df['trip_duration'], 99.7)]
train_df = train_df[(train_df['total_distance']>= distLimits[0]) & (train_df['total_distance'] <= distLimits[1])]
train_df = train_df[(train_df['trip_duration']>= durLimits[0]) & (train_df['trip_duration'] <= durLimits[1])]
```

```python
sns.set(style ='white')

fig = plt.figure(constrained_layout = True,figsize = (12,8))
gs = GridSpec(2,3,figure=fig)

ax1 = fig.add_subplot(gs[0,1:])
ax2 = fig.add_subplot(gs[1,1:])
ax3 = fig.add_subplot(gs[0:,0])
ax3.axis('off')


sns.distplot(train_df['total_distance']/1000,ax = ax1, axlabel = "",color='lightgreen')
ax1.set_title("TOTAL DISTANCE in Km.",size = 20, color = 'grey',fontfamily ='Franklin Gothic Medium',loc ='right')
ax1.tick_params(labelleft = False )

sns.distplot(train_df['trip_duration']/60.0,ax = ax2, color = 'green',axlabel = "")
ax2.set_title("TRIP DURATION in Min.",size = 20, color = 'grey',fontfamily ='Franklin Gothic Medium',loc='right')
ax2.tick_params(labelleft = False)

ax3.text(0.1,0.75,'Most trips are less than \n ',
         fontsize = 20, ha = 'center', color='grey', fontfamily ='Franklin Gothic Medium')
ax3.text(0.3,0.7,'6 KMS',
         fontsize = 40, ha = 'center', color='mediumseagreen', fontfamily ='Franklin Gothic Medium')
ax3.text(0.1,0.3,'Trips usually last for less than',
         fontsize = 20, ha = 'center', color='grey',fontfamily ='Franklin Gothic Medium')
ax3.text(0.3,0.20,'20 MINS',
         fontsize = 40, ha = 'center', color='mediumseagreen',fontfamily ='Franklin Gothic Medium')
```

```python
train_df['pickup_datetime'] = pd.to_datetime(train_df.pickup_datetime)
train_df['dropoff_datetime'] = pd.to_datetime(train_df.dropoff_datetime)
test_df['pickup_datetime']= pd.to_datetime(test_df.pickup_datetime)
```

```python
for df in (train_df,test_df):
    df['pickupdate'] = df['pickup_datetime'].dt.date
    df['pickupday'] = df['pickup_datetime'].dt.day
    df['pickupmonth'] = df['pickup_datetime'].dt.month_name()
    df['pickupweekday'] = df['pickup_datetime'].dt.day_name()
    df['pickuphour']= df['pickup_datetime'].dt.hour
```

```python
hr_day_dur_df = pd.DataFrame(train_df.groupby(['pickupweekday','pickuphour'])['DurationMIN'].agg('mean'))
hr_day_dur_df.reset_index(inplace =True)
```

```python
#hr vs duration weekday 
fig = px.bar(hr_day_dur_df,x='pickuphour',y='DurationMIN',
             facet_row='pickupweekday',color='DurationMIN',
             color_continuous_scale=px.colors.sequential.Mint,
             text='DurationMIN',height =800,width=1000,
             title=dict(text="Avg. Trip Duration During Different Hours of the Day, for Every Day of the Week ",
                        font_family ="Franklin Gothic",
                        font_size=25,
                        x=0.5),
             category_orders=dict(pickupweekday=['Monday','Tuesday','Wednesday',
                                                 'Thursday','Friday',
                                                 'Saturday','Sunday']))


fig.update_xaxes(row =1,title ="HOUR OF THE DAY")


fig.update_yaxes(row =4,title ="IN")
fig.update_yaxes(row =3,title ="DURATION")
fig.update_yaxes(row =5,title ="MINUTES")

fig.update_yaxes(row =2,title ="")
fig.update_yaxes(row =7,title ="")
fig.update_yaxes(row =6,title ="")
fig.update_yaxes(row =1,title ="")


fig.for_each_annotation(lambda x: x.update(text = x.text.split('=')[-1][0:3]))

fig.update_traces(texttemplate='%{text:.2s}', textposition='inside')
fig.update_traces(marker_line_color='black',
                  marker_line_width=2, opacity=0.7)
fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)

fig.add_annotation(row=2,col=0, x = 8, y=9, text = 'Dip in duration because ..weekend?',arrowcolor = 'black', arrowhead=7)
fig.add_annotation(row=1,col=0, x = 8, y=9, text = 'Probably.',arrowcolor = 'black', arrowhead=7)
```

```python
day_month_dur_df = pd.DataFrame(train_df.groupby(['pickupmonth','pickupday'])['DurationMIN'].agg('mean'))
day_month_dur_df.reset_index(inplace =True)
```

```python
fig = px.bar(day_month_dur_df, 
             x ='pickupmonth',
             y='DurationMIN',
             color='DurationMIN',
             color_continuous_scale=py.colors.sequential.PuRd,
             hover_data=['pickupday'],
             text='DurationMIN',
             title=dict(text="Avg. Trip Duration for Different Days of the Month",
                        font_family ="Franklin Gothic Medium",
                        font_size=30,
                        x=0.5),
             category_orders=dict(pickupmonth =['January','February','March','April','May','June']),
             height=800)

fig.update_traces(texttemplate='%{text:.2s}', textposition='inside')
fig.update_traces(marker_line_color='black',
                  marker_line_width=2, opacity=0.8)

fig.update_yaxes(title ='DURATION  IN  MINUTES')
fig.update_xaxes(title ='MONTH')
fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
```

```python
vendor_dur_df = pd.DataFrame(train_df.groupby(['vendor_id','pickupdate'])['DurationMIN'].mean())
vendor_dur_df.reset_index(inplace =True)
```

```python
fig = px.line(vendor_dur_df,x ='pickupdate',y='DurationMIN',color ='vendor_id',
             title=dict(text="Avg. Trip Duration for Each Vendor ",
                        font_family ="Franklin Gothic Medium",
                        font_size=30,
                        x=0.5))
fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(title ='DURATION  IN  MINUTES')
fig.update_xaxes(title ='DATE')
```

```python
west, south, east, north = -74.03, 40.63, -73.77, 40.85

train_df = train_df[(train_df.pickup_latitude> south) & (train_df.pickup_latitude < north)]
train_df = train_df[(train_df.dropoff_latitude> south) & (train_df.dropoff_latitude < north)]
train_df = train_df[(train_df.pickup_longitude> west) & (train_df.pickup_longitude < east)]
train_df = train_df[(train_df.dropoff_longitude> west) & (train_df.dropoff_longitude < east)]
```

```python
fig = plt.figure(constrained_layout=True,figsize=(12,8))
gs = GridSpec(1,3,figure = fig)
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[0,2])
ax1.axis('off')
ax2.set_facecolor('grey')
ax2.set()
ax3.set_facecolor('grey')
sns.scatterplot(data=train_df ,x ='pickup_longitude', y='pickup_latitude',ax=ax2, s=.02, alpha=.6,)
sns.scatterplot(data=train_df, x ='dropoff_longitude', y='dropoff_latitude',ax=ax3, s=.02, alpha=.6,)
ax1.text(0,0.6,
         'Pickup locations are \nclustered around \nManhattan, \nJohn F. Kennedy, \nLA Guardia Airport.',
         size =20)
ax1.text(0,0.2,
         'Drop off locations are \ncompartively more \nwidespread, though \nclustered around the \nabove mention places.',
         size =20)
```
